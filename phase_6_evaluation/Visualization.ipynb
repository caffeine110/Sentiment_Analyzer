```python
kk = []
for i in range(10):
    kk.append(yy[i][0])
    del(yy[i][0])
mm = list(map(list, zip(*yy)))
```

```python

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
 
# Plot
plt.stackplot(kk,mm[1:3], labels=['No','Yes','Total'])
plt.legend(loc='upper left')
plt.xticks(rotation=90)
plt.show()
```

```python
# Data to plot
labels = 'No','Yes'
colors = ['lightcoral', 'lightskyblue']
explode = (0.1, 0)  # explode 1st slice
 
# Plot
for i in range(10):
    plt.pie(yy[i][1:3], explode=explode, labels=labels, colors=colors,
    autopct='%1.1f%%', shadow=True, startangle=140)
    plt.title(kk[i])
    plt.axis('equal')
    plt.show()
```
